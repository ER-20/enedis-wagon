```python
# Create a dictionary that maps DPE values to integers
dpe_mapping = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5, 'F': 6, 'G': 7}

# Create a new column 'DPE_numeric' that maps the DPE values to integers
df_log_existants_2022['DPE_numeric'] = df_log_existants_2022['Etiquette_DPE'].map(dpe_mapping)
```

```python
df_log_existants['Date_réception_DPE'] = pd.to_datetime(df_log_existants['Date_réception_DPE'], errors='coerce')
```

```python
df_log_existants['Year_reception_DPE'] = df_log_existants['Date_réception_DPE'].dt.year
```

```python
df_log_existants_2022 = df_log_existants[df_log_existants['Year_reception_DPE'] == 2022]
```

```python
# Display the first few rows of the DataFrame, including the 'Etiquette_DPE' and 'DPE_numeric' columns
df_log_existants_2022[['Etiquette_DPE', 'DPE_numeric']].head()
```

```python
# Create a new DataFrame with the 'DPE_numeric', 'Etiquette_DPE', and 'Identifiant__BAN' columns
df_simple_2022 = pd.DataFrame({
    'DPE_numeric': df_log_existants_2022['DPE_numeric'],
    'Etiquette_DPE': df_log_existants_2022['Etiquette_DPE'],
    'Identifiant__BAN': df_log_existants_2022['Identifiant__BAN']
})
```

```python
# Calculate the average DPE per id_ban and round it to the nearest integer
df_average_dpe_per_id_ban_2022 = df_simple_2022.groupby('Identifiant__BAN')['DPE_numeric'].mean().round().reset_index()
```

```python
df_average_dpe_per_id_ban_2022.shape
```

```python
df_average_dpe_per_id_ban_2022.head()
```

```python
# Create a reverse mapping dictionary that maps integers to DPE values
reverse_dpe_mapping = {1.0: 'A', 2.0: 'B', 3.0: 'C', 4.0: 'D', 5.0: 'E', 6.0: 'F', 7.0: 'G'}

# Create a new column 'Etiquette_DPE_rounded' that maps the rounded DPE values to their corresponding letters
df_average_dpe_per_id_ban_2022['Etiquette_DPE_per_id_ban'] = df_average_dpe_per_id_ban_2022['DPE_numeric'].map(reverse_dpe_mapping)
```
